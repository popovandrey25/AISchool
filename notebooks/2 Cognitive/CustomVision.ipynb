Создаём проект:

```python
from azure.cognitiveservices.vision.customvision.training import custom_vision_training_client
from azure.cognitiveservices.vision.customvision.training.models import ImageFileCreateEntry

trainer = custom_vision_training_client.CustomVisionTrainingClient(training_key,"https://southcentralus.api.cognitive.microsoft.com")

# Create a new project
print ("Creating project...")
project = trainer.create_project("CatsDogs")
project_id = project.id
print("Project id={} created".format(project_id))
```

Создаём теги для кошек и собак:

```python
cat_tag = trainer.create_tag(project_id, "Cat")
dog_tag = trainer.create_tag(project_id, "Dog")
```

Загружаем наши обучающие изображения в сервис:

```python
for fn in os.listdir("train_set"):
    print("Uploading %s" % fn)
    with open(os.path.join("train_set",fn),'rb') as f:
        data = f.read()
        trainer.create_images_from_data(project_id,tag_ids=[cat_tag.id if "cat" in fn else dog_tag.id],image_data=data)
```

Прежде чем использовать проект, его необходимо обучить:

```python
import time

print ("Training...")
iteration = trainer.train_project(project.id)
while (iteration.status != "Completed"):
    iteration = trainer.get_iteration(project.id, iteration.id)
    print ("Training status: " + iteration.status)
    time.sleep(1)

# The iteration is now trained. Make it the default project endpoint
trainer.update_iteration(project.id, iteration.id, is_default=True)
print ("Done!")
```

## Часть 2: Проверяем точность модели

Вы можете посмотреть точность модели на обучающих данных через веб-интерфейс. Однако интерес представляет проверка модели на тестовых данных.

Вы можете выбрать два способа проверки:

* Вручную через веб-интерфейс, загрузив несколько файлов и посчитав примерный процент правильных распознаваний
* С помощью автоматического скрипта ниже

Вы можете использовать скрипт даже в том случае, если загружали файлы вручную - в этом случае в переменную `project_id` надо записать идентификатор вашего проекта, взяв его из веб-интерфейса.

Для начала подготовим случайные файлы из нашего набора для тестирования:

```python
def getfiles(cls,n):
    files = os.listdir(os.path.join("test_set",cls))
    random.shuffle(files)
    return [os.path.join("test_set",cls,k) for k in files[0:n]]

test_set = getfiles('cats',n_test) + getfiles('dogs',n_test)
```

Для использование сервиса создаём объект `CustomVisionPredictionClient`, указывая `prediction_key`

```python
from azure.cognitiveservices.vision.customvision.prediction import custom_vision_prediction_client
from azure.cognitiveservices.vision.customvision.prediction import models

predictor = custom_vision_prediction_client.CustomVisionPredictionClient(prediction_key,"https://southcentralus.api.cognitive.microsoft.com")
```

Для обработки изображения нашей моделью создадим функцию `predict`, которая будет принимать имя файла, передавать его сервису, и возвращать вероятности принадлежности картинки к тому или иному тегу:

```python
def predict(fn):
    with open(fn,'rb') as f:
        data = f.read()
        return predictor.predict_image(project_id,data)

res = predict(test_set[0])
# Display the results.
for prediction in res.predictions:
    print ("\t" + prediction.tag_name + ": {0:.2f}%".format(prediction.probability * 100))
```

Наконец, получим точность модели, вызвав функцию `predict` на всех файлах текстовой выборки:

```python
total = 0
correct = 0
for x in test_set:
    res = predict(x)
    if res.predictions[0].tag_name.lower() in x:
        correct+=1
    total+=1
print("Accuracy: {}%".format(100*correct/total))
```
